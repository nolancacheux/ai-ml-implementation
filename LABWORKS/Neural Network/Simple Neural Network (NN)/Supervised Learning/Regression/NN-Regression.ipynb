```markdown
# Neural Network Model for Regression

This Jupyter Notebook demonstrates the creation, training, and evaluation of a neural network model for regression using TensorFlow and Keras. The model is trained on synthetic data and its performance is visualized through plots.
```

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense

model=Sequential()
model.add(Input(shape=(5,)))
model.add(Dense(6))
model.add(Dense(7))
model.add(Dense(7))
model.add(Dense(6))
model.add(Dense(5))

model.summary()

````

````python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt

# Generate synthetic data
x = np.linspace(0, 1, 100)
y = 2 * x + np.random.normal(0, 0.1, x.shape) + 1

# Define the neural network model
model = Sequential()
model.add(Dense(1000, input_dim=1, activation='linear'))
model.add(Dense(1000, activation='linear'))
model.add(Dense(1))

# Compile the model
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.1), loss='mean_squared_error')

# Train the model
model.fit(x, y, epochs=100, batch_size=10)
# Predict using the model
predictions = model.predict(x)

model.summary()

plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Original Data')
plt.plot(x, predictions, color='red', label='Predictions')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Original Data vs Predictions')
plt.legend()
plt.show()
````

Qu'est-ce que l'activation ReLU ? (= Rampe en maths)
ReLU, ou Rectified Linear Unit, est une fonction d'activation couramment utilisée dans les réseaux de neurones. Elle est définie par la formule suivante : 
ReLU}(x) = max(0, x)

Intérêt de l'activation ReLU
- **Non-linéarité** : ReLU introduit de la non-linéarité dans le modèle. Cela permet au réseau de neurones d'apprendre des relations complexes dans les données, ce qui est crucial pour traiter des problèmes réels souvent non linéaires.

- **Simplicité et efficacité** : La fonction ReLU est simple à calculer car elle consiste simplement à prendre le maximum entre zéro et l'entrée. Cela la rend plus rapide à évaluer que d'autres fonctions d'activation comme la sigmoïde ou la tangente hyperbolique.

- **Activation éparse** : ReLU peut entraîner une activation éparse, ce qui signifie que dans une couche donnée, certains neurones auront une sortie de zéro. Cela peut rendre le réseau plus efficace et réduire le risque de surapprentissage (overfitting).

- **Propagation du gradient** : Contrairement à la sigmoïde ou à la tangente hyperbolique, ReLU ne souffre pas du problème de gradient qui disparaît pour les entrées positives. Cela aide à propager efficacement les gradients pendant la rétropropagation, ce qui est essentiel pour l'entraînement des réseaux profonds.

````python
# Generate new synthetic data using sine function
y_sine = np.sin(2 * np.pi * x)

# Define the neural network model
model_sine = Sequential()
model_sine.add(Dense(1000, input_dim=1, activation='relu'))
model_sine.add(Dense(1000, activation='relu'))
model_sine.add(Dense(1))

# Compile the model
model_sine.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.1), loss='mean_squared_error')

# Train the model
model_sine.fit(x, y_sine, epochs=100, batch_size=10)

# Predict using the model
predictions_sine = model_sine.predict(x)

plt.figure(figsize=(10, 6))
plt.scatter(x, y_sine, label='Original Data (Sine)')
plt.plot(x, predictions_sine, color='red', label='Predictions')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Original Data (Sine) vs Predictions')
plt.legend()
plt.show()
````
